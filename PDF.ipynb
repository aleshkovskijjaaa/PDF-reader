```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text += page.get_text()
    return text

# Пример использования
pdf_path = "metodika-diagnostiki-tipa-vybora-zhiznennogo-puti-lichnosti.pdf"
text = extract_text_from_pdf(pdf_path)
print(text)
```

```python
import pdfplumber

def extract_tables_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        tables = []
        for page in pdf.pages:
            extracted_tables = page.extract_tables()
            tables.extend(extracted_tables)
    return tables

# Пример использования
pdf_path = "metodika-diagnostiki-tipa-vybora-zhiznennogo-puti-lichnosti.pdf"
tables = extract_tables_from_pdf(pdf_path)
for table in tables:
    for row in table:
        print(row)
```

```python
from pdfminer.high_level import extract_text

def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

def parse_tables_from_text(text):
    lines = text.split('\n')
    tables = []
    table = []

    for line in lines:
        # Предположим, что строки таблицы содержат числа и разделены пробелами
        if any(char.isdigit() for char in line):
            cells = line.split()
            table.append(cells)
        else:
            if table:
                tables.append(table)
                table = []

    if table:
        tables.append(table)

    return tables

# Пример использования
pdf_path = "metodika-diagnostiki-tipa-vybora-zhiznennogo-puti-lichnosti.pdf"
text = extract_text_from_pdf(pdf_path)
tables = parse_tables_from_text(text)

print("Extracted Tables:")
for table in tables:
    for row in table:
        print(row)
```

```python
import PyPDF2
import pandas as pd
import numpy as np
from scipy import stats

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text

def analyze_data(data):
    df = pd.DataFrame(data)
    mean_values = df.mean()
    p_values = df.apply(lambda x: stats.ttest_1samp(x.dropna(), 0).pvalue)
    return mean_values, p_values

# Пример использования
pdf_path = "metodika-diagnostiki-tipa-vybora-zhiznennogo-puti-lichnosti.pdf"
text = extract_text_from_pdf(pdf_path)

# Для демонстрации создадим данные вручную
data = [
    [1.2, 2.3, 3.1],
    [2.1, 3.4, 1.2],
    [0.9, 1.2, 2.8]
]
mean_values, p_values = analyze_data(data)

print("Extracted Text:")
print(text)
print("\nMean Values:")
print(mean_values)
print("P-Values:")
print(p_values)
```
